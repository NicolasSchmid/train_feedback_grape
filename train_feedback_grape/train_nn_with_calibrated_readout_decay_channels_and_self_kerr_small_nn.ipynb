## Driven Hamiltonian

```python
def build_drive_htree(N_cav, chi, e_qub, e_cav):
    a = tensor(identity(2), destroy(N_cav))
    adag = hconj(a)
    n_phot = adag @ a
    sigz = tensor(sigma.z, identity(N_cav))
    sigp = tensor(sigma.p, identity(N_cav))
    one = tensor(identity(2), identity(N_cav))

    H_disp = (chi / 2) * n_phot @ (one - sigz)
    H_kerr = 0.5 * K_c * adag @ adag @ a @ a

    # Combined drift Hamiltonian
    H_drift = H_disp + H_kerr
    
    ht = [
        [H_drift, 1, 1, 0],
        [mu_qub * sigp, e_qub, 1, 1],                
        [mu_cav * adag, 1j *jnp.conj(e_cav), 1, 1] 
    ]
    return ht

def get_uevol_matrix(N_cav, chi, e_qub, e_cav, dt):
    htree = build_drive_htree(N_cav, chi, e_qub, e_cav)
    hmat = hmat_from_htree(htree)
    return evol_hdt_exp(hmat, dt)

def make_drive(N_cav, chi):
    
    def drive(rho_init, es_qub, es_cav, dts):
        
        def f(rho, x):
            e_qub, e_cav, dt = x
            uevol = get_uevol_matrix(N_cav, chi, e_qub, e_cav, dt)[0]
            return uevol @ rho @ hconj(uevol), None
    
        rho_fin, _ = jax.lax.scan(f, rho_init, [es_qub, es_cav, dts])
        return rho_fin
    
    return drive

```

## Relaxation

```python
def qubit_ops():
    g = basis(2)       # |g> = index 0
    e = basis(2, 1)    # |e> = index 1

    Pg = g @ hconj(g)
    Pe = e @ hconj(e)

    sm = g @ hconj(e)  # |g><e|, relaxation operator
    sp = e @ hconj(g)  # |e><g|, excitation operator

    sz = Pg - Pe       # diag(+1, -1)

    return g, e, Pg, Pe, sm, sp, sz

def liouv_relax_channel(c):
    l = tensor(c, c.conj())
    l -= 1 / 2 * tensor(hconj(c) @ c, identity(len(c)))
    l -= 1 / 2 * tensor(identity(len(c)), (hconj(c) @ c).T)
    return l

def eval_idling_matrix(cs, t):
    l = sum(liouv_relax_channel(c) for c in cs)
    return jsp.linalg.expm(l * t)

def build_c_ops(N_cav, t1_qub, t2_qub, t1_cav):
    get_t_phi_qub = lambda t1, t2: 1 / ((1 / t2) - (1 / t1) / 2)
    t_phi_qub = get_t_phi_qub(t1_qub, t2_qub)

    g, e, Pg, Pe, sm, sp, sz = qubit_ops()

    sig_down = tensor(sm, identity(N_cav))
    sigz = tensor(sz, identity(N_cav))
    a = tensor(identity(2), destroy(N_cav))
    
    c_ops = [
        1 / jnp.sqrt(t1_qub) * sig_down,
        1 / jnp.sqrt(t_phi_qub) * sigz / jnp.sqrt(2),
        1 / jnp.sqrt(t1_cav) * a,
    ]

    return c_ops

def make_idle(N_cav, t1_qub, t2_qub, t1_cav, t):
    cs = build_c_ops(N_cav, t1_qub, t2_qub, t1_cav)
    idling_matrix = eval_idling_matrix(cs, t)

    def idle(rho):
        rho_liouv = rho.reshape(-1, 1)
        rho_liouv = idling_matrix @ rho_liouv
        return rho_liouv.reshape(rho.shape)

    return idle


print("sigma.m =", sigma.m)
```

# Control with interleaved decay

```python
def make_drive_with_interleaved_decay(
    N_cav,
    chi,
    t1_qub,
    t2_qub,
    t1_cav,
    t_control=1.408,      # us
    n_decay_chunks=4,     # start with 4 or 8
):
    """
    Approximate dissipative evolution during the feedback-GRAPE pulse.

    Instead of:
        full unitary drive, then full idle

    use Strang splitting:
        half idle -> unitary chunk -> half idle
        repeated over chunks.

    This is a better approximation when the qubit can be populated during
    the control pulse.
    """

    drive = make_drive(N_cav, chi)

    n_intvs = len(t_drive_intvs)

    # Fixed Python slices, so JAX sees a static unrolled loop.
    edges = np.linspace(0, n_intvs, n_decay_chunks + 1, dtype=int)
    chunk_slices = [(int(edges[i]), int(edges[i + 1])) for i in range(n_decay_chunks)]

    dt_chunk = t_control / n_decay_chunks

    # Half-step dissipative evolution for Strang splitting.
    idle_half_chunk = make_idle(
        N_cav,
        t1_qub,
        t2_qub,
        t1_cav,
        0.5 * dt_chunk,
    )

    def drive_with_decay(rho, es_qub, es_cav):
        for start, stop in chunk_slices:
            if stop > start:
                # Half dissipative step before this unitary chunk
                rho = idle_half_chunk(rho)

                # Unitary control chunk
                rho = drive(
                    rho,
                    es_qub[start:stop],
                    es_cav[start:stop],
                    t_drive_intvs[start:stop],
                )

                # Half dissipative step after this unitary chunk
                rho = idle_half_chunk(rho)

        return rho

    return drive_with_decay
```

## Fidelity of the cavity state

```python
def fidelity_cav(rho,psi_target_cav):
    idx_mid = rho.shape[-1] // 2
    rho_cav = rho[:idx_mid, :idx_mid] + rho[idx_mid:, idx_mid:]
    return (hconj(psi_target_cav) @ rho_cav @ psi_target_cav)[0][0].real

def fidelity_cav_only_cav_rho(rho,psi_target_cav):
    return (hconj(psi_target_cav) @ rho @ psi_target_cav)[0][0].real
```

## Measurement with qubit reset

```python
import jax
import jax.numpy as jnp
import jax.scipy.linalg as jsp_linalg


# ============================================================
# Effective cavity-only channels used inside the readout model
# ============================================================

def make_cavity_loss_only(N_cav, t1_cav, t):
    """
    Cavity amplitude damping during time t.

    Units:
        t and t1_cav must be in the same units, here us.

    This uses the same Liouville convention as the rest of the notebook.
    """
    a = destroy(N_cav)
    c = (1.0 / jnp.sqrt(t1_cav)) * a

    L = liouv_relax_channel(c)
    M = jsp_linalg.expm(L * t)

    def cavity_loss(rho_cav):
        rho_vec = rho_cav.reshape(-1, 1)
        rho_vec = M @ rho_vec
        return rho_vec.reshape(rho_cav.shape)

    return cavity_loss


def apply_phase_rotation_and_diffusion(rho_cav, theta_rad, sigma_theta_rad):
    """
    Averaged random phase-space rotation.

    Deterministic part:
        rho_nm -> exp(i theta (n-m)) rho_nm

    Gaussian shot-to-shot phase uncertainty:
        rho_nm -> exp[-0.5 sigma_theta^2 (n-m)^2] rho_nm
    """
    N = rho_cav.shape[0]
    n = jnp.arange(N)

    dn = n[:, None] - n[None, :]

    phase = jnp.exp(1j * theta_rad * dn)
    deph = jnp.exp(-0.5 * (sigma_theta_rad ** 2) * (dn ** 2))

    return rho_cav * phase * deph


# ============================================================
# Effective few-channel readout model
# ============================================================

def make_effective_readout(
    N_cav,
    t1_qub,
    t1_cav,
    psi_target_cav,
    *,
    readout_duration,       # us

    # Mean readout-induced cavity rotations
    theta_g_deg,
    theta_e_deg,

    # Readout-induced phase diffusion
    sigma_g_deg,
    sigma_e_deg,

    # Readout assignment errors
    # p_g_to_e = P(record excited | qubit effectively ground)
    # p_e_to_g = P(record ground   | qubit effectively excited)
    p_g_to_e,
    p_e_to_g,

    # Number of effective decay-time bins for e -> g during readout
    n_jump_bins,
):
    """
    Effective measurement + readout backaction model.

    Output convention:
        recorded ground  -> meas_out = +1
        recorded excited -> meas_out = -1

    Physics included:
        1. Projective qubit measurement into rho_gg / rho_ee blocks.
        2. Cavity loss during the readout.
        3. Readout-induced cavity rotation.
        4. Readout-induced phase diffusion.
        5. Qubit T1 decay during readout using a few hidden jump-time bins.
        6. Classical readout assignment errors.

    Important:
        The sampled variable is still only the recorded bit.
        The hidden decay branches are summed analytically.
    """

    cavity_loss_ro = make_cavity_loss_only(N_cav, t1_cav, readout_duration)

    theta_g = jnp.deg2rad(theta_g_deg)
    theta_e = jnp.deg2rad(theta_e_deg)

    sigma_g = jnp.deg2rad(sigma_g_deg)
    sigma_e = jnp.deg2rad(sigma_e_deg)

    # ------------------------------------------------------------
    # Effective excited-state decay branches during readout
    # ------------------------------------------------------------
    #
    # Branch 0:
    #     no jump during the readout
    #
    # Branch 1..n_jump_bins:
    #     jump happens inside a time bin.
    #
    # If the jump happens after fraction f of the readout,
    # approximate the cavity rotation as
    #
    #     theta_branch = f theta_e + (1 - f) theta_g
    #
    # The uncertainty sigma is interpolated in the same way:
    #
    #     sigma_branch = f sigma_e + (1 - f) sigma_g
    #
    # There is no separate sigma_jump.
    # ------------------------------------------------------------

    edges = jnp.linspace(0.0, 1.0, n_jump_bins + 1)
    mids = 0.5 * (edges[:-1] + edges[1:])

    gammaT = readout_duration / t1_qub

    p_no_jump = jnp.exp(-gammaT)
    p_jump_bins = jnp.exp(-gammaT * edges[:-1]) - jnp.exp(-gammaT * edges[1:])

    frac_excited = jnp.concatenate([
        jnp.array([1.0]),  # no jump
        mids,
    ])

    e_branch_weights = jnp.concatenate([
        jnp.array([p_no_jump]),
        p_jump_bins,
    ])

    # Numerical safety
    e_branch_weights = e_branch_weights / jnp.sum(e_branch_weights)

    e_branch_thetas = frac_excited * theta_e + (1.0 - frac_excited) * theta_g
    e_branch_sigmas = frac_excited * sigma_e + (1.0 - frac_excited) * sigma_g

    # Effective probability of recording excited for an excited-origin branch.
    # If f=0, branch looks like ground readout.
    # If f=1, branch looks like excited readout.
    p_rec_e_for_e_branch = (
        p_g_to_e
        + frac_excited * ((1.0 - p_e_to_g) - p_g_to_e)
    )
    p_rec_g_for_e_branch = 1.0 - p_rec_e_for_e_branch

    # Ground-origin branch assignment probabilities.
    p_rec_g_for_g_branch = 1.0 - p_g_to_e
    p_rec_e_for_g_branch = p_g_to_e

    def branch_channel(rho_cav_unnorm, theta, sigma):
        """
        Linear channel on an unnormalized cavity block.
        """
        rho = cavity_loss_ro(rho_cav_unnorm)
        rho = apply_phase_rotation_and_diffusion(rho, theta, sigma)
        return rho

    def readout(key, rho_init):
        """
        Effective readout step.

        Input:
            rho_init:
                full qubit+cavity density matrix,
                shape (2*N_cav, 2*N_cav)

        Output:
            meas_out:
                +1 for recorded ground, -1 for recorded excited

            prob:
                probability of the sampled recorded outcome

            rho_fin:
                full qubit+cavity density matrix after readout
                and qubit reset to ground

            fid_after_readout:
                diagnostic cavity fidelity after readout backaction
        """

        # Qubit blocks. The qubit coherences are discarded by the measurement.
        rho_gg = rho_init[:N_cav, :N_cav]
        rho_ee = rho_init[N_cav:, N_cav:]

        # --------------------------
        # Ground-origin hidden branch
        # --------------------------
        rho_g_branch = branch_channel(rho_gg, theta_g, sigma_g)

        rho_record_g = p_rec_g_for_g_branch * rho_g_branch
        rho_record_e = p_rec_e_for_g_branch * rho_g_branch

        # ---------------------------
        # Excited-origin hidden branches
        # ---------------------------
        for i in range(n_jump_bins + 1):
            rho_e_i = branch_channel(
                rho_ee,
                e_branch_thetas[i],
                e_branch_sigmas[i],
            )

            rho_e_i = e_branch_weights[i] * rho_e_i

            rho_record_g = rho_record_g + p_rec_g_for_e_branch[i] * rho_e_i
            rho_record_e = rho_record_e + p_rec_e_for_e_branch[i] * rho_e_i

        # Probabilities of the recorded classical outcomes
        p_g_raw = jnp.real(jnp.trace(rho_record_g))
        p_e_raw = jnp.real(jnp.trace(rho_record_e))

        p_raw = jnp.array([p_g_raw, p_e_raw])
        p_raw = jnp.maximum(p_raw, 0.0)

        p_sample = p_raw / jnp.sum(p_raw)

        outcome_indx = jax.random.choice(key, 2, p=p_sample)
        prob = p_sample[outcome_indx]

        # Conditional cavity state given the recorded bit
        eps = 1e-12

        rho_cav_g = rho_record_g / (p_g_raw + eps)
        rho_cav_e = rho_record_e / (p_e_raw + eps)

        rho_cav = jnp.where(
            outcome_indx == 0,
            rho_cav_g,
            rho_cav_e,
        )

        rho_cav = rho_cav / jnp.trace(rho_cav)

        # Diagnostic only: not necessarily the training reward
        fid_after_readout = fidelity_cav_only_cav_rho(rho_cav, psi_target_cav)

        # Reset qubit to |g>
        rho_qub_g = basis(2) @ hconj(basis(2))
        rho_fin = tensor(rho_qub_g, rho_cav)
        rho_fin = rho_fin / jnp.trace(rho_fin)

        # recorded ground -> +1
        # recorded excited -> -1
        meas_out = 1 - 2 * outcome_indx

        return meas_out, prob, rho_fin, fid_after_readout

    return readout
```

## RNN

```python
def init_carry(key, rec_features, inp_shape):
    gru_cell = nn.GRUCell(features=rec_features)
    return gru_cell.initialize_carry(key, inp_shape)

def create_flax_state(key,
                      rec_features, dense_features,
                      inp_shape, learning_rate,
                      *, print_summary):
    
    class Model(nn.Module):
    
        @nn.compact
        def __call__(self, carry, x):
            
            gru_cell = nn.GRUCell(features=rec_features)
            carry, x = gru_cell(carry, x)
            
            for ndf in dense_features:
                x = nn.Dense(ndf)(x)
                x = jax.nn.relu(x)
                
            out_layer = nn.Dense(4 * num_bspln)
            x = out_layer(x)
            x = x.reshape(4, num_bspln)
            
            return carry, x

    model = Model()

    fake_key = jax.random.key(0)
    fake_carry = init_carry(fake_key, rec_features, inp_shape)
    fake_inp = jnp.ones(inp_shape)
    params = model.init(key, fake_carry, fake_inp)["params"]

    # optimizer = optax.adam(learning_rate=learning_rate)
    optimizer = optax.chain( #avoid unlikely trajectory to create catastrophic update
    optax.clip_by_global_norm(1.0),
    optax.adam(learning_rate=learning_rate),
    )

    flax_state = TrainState.create(apply_fn=model.apply, params=params, tx=optimizer)

    if print_summary:
        print(model.tabulate(fake_key, fake_carry, fake_inp))

    return flax_state

def apply_nn(flax_state, params, carry, inp):
    return flax_state.apply_fn({'params': params}, carry, inp)

```

## Time step  
- 1) Measurement
- 2) Fidelity
- 3) Idling (decay)
- 3) RNN interrogation
- 3) Qubit reset

```python
def make_time_step(
    N_cav,
    chi,
    t1_qub,
    t2_qub,
    t1_cav,
    t_idle,
    psi_target_cav,
    t_control=1.408,       # us
    n_decay_chunks=4,
):
    readout = make_effective_readout(
        N_cav=N_cav,
        t1_qub=t1_qub,
        t1_cav=t1_cav,
        psi_target_cav=psi_target_cav,

        readout_duration=1.6+0.4, #readout +feedback time

        theta_g_deg=-18.67,
        theta_e_deg=+13.78,

        sigma_g_deg=0.5,
        sigma_e_deg=2.0,

        p_g_to_e=0.00,
        p_e_to_g=0.04,

        n_jump_bins=5,
    )

    idle = make_idle(N_cav, t1_qub, t2_qub, t1_cav, t_idle)

    drive_with_decay = make_drive_with_interleaved_decay(
        N_cav=N_cav,
        chi=chi,
        t1_qub=t1_qub,
        t2_qub=t2_qub,
        t1_cav=t1_cav,
        t_control=t_control,
        n_decay_chunks=n_decay_chunks,
    )

    def time_step(key, rho, flax_state, params, carry):
        # 1. Measurement + effective readout backaction
        meas_out, prob, rho, fid_after_readout = readout(key, rho)

        # 2. Idle / feedback latency
        rho = idle(rho)

        # 3. RNN receives only the recorded bit
        rnn_inp = jnp.array([meas_out], dtype=float)
        carry, ctrl_coeffs = apply_nn(flax_state, params, carry, rnn_inp)

        # 4. Apply feedback-GRAPE pulse with interleaved decoherence
        es_qub, es_cav = ctrl_from_coeffs(ctrl_coeffs, bsplns_mids)
        rho = drive_with_decay(rho, es_qub, es_cav)

        # 5. Training fidelity after control and control-pulse decoherence
        fid_after_control = fidelity_cav(rho, psi_target_cav)

        return rho, carry, fid_after_control, meas_out, prob

    return time_step
```

## Feedback-GRAPE simulation model

At each feedback step, the input state is the full qubit+cavity density matrix $\rho_k$.  
One simulation step is modeled as

$$
\rho_k
\;\xrightarrow{\mathrm{effective\ readout}}\;
\rho_k^{(\mathrm{ro},m_k)}
\;\xrightarrow{\mathrm{idle}}\;
\rho_k^{(\mathrm{idle},m_k)}
\;\xrightarrow{\mathrm{RNN\ control+diss.}}\;
\rho_{k+1}.
$$

The recorded readout bit is

$$
m_k =
\begin{cases}
+1, & \mathrm{recorded\ ground},\\
-1, & \mathrm{recorded\ excited}.
\end{cases}
$$

Only this recorded bit $m_k$ is given to the RNN.

---

### 1. Effective qubit readout model used in the simulation

At each feedback step, the input state is the full qubit+cavity density matrix $\rho$.  
We model the readout as a measurement of the qubit, plus an effective backaction on the cavity.



#### 1. Qubit projection: split the state into ground/excited blocks

In the basis $\{|g\rangle, |e\rangle\}$, the full density matrix is written as

$$
\rho =
\begin{pmatrix}
\rho_{gg} & \rho_{ge} \\
\rho_{eg} & \rho_{ee}
\end{pmatrix}.
$$

The readout destroys the qubit coherences, so we keep only the two cavity blocks

$$
\rho_{gg}, \qquad \rho_{ee}.
$$

Their traces are the probabilities that the qubit was effectively in $|g\rangle$ or $|e\rangle$ before the readout:

$$
p_g = \mathrm{Tr}(\rho_{gg}), 
\qquad
p_e = \mathrm{Tr}(\rho_{ee}).
$$



#### 2. Cavity backaction during the readout

During the readout, the cavity state is modified in three ways:

1. cavity decay during the readout time $T_\mathrm{ro}$,
2. a readout-induced phase-space rotation,
3. readout-induced phase diffusion.

For a cavity block $\rho_\mathrm{cav}$, the effective channel is

$$
\rho_\mathrm{cav}
\longmapsto
\mathcal{R}_{\theta,\sigma}
\left[
\mathcal{L}_{T_\mathrm{ro}}(\rho_\mathrm{cav})
\right],
$$

where $\mathcal{L}_{T_\mathrm{ro}}$ is cavity loss during the readout.

The rotation plus phase diffusion is applied in the Fock basis as

$$
\left[\mathcal{R}_{\theta,\sigma}(\rho)\right]_{nm}
=
e^{i\theta(n-m)}
e^{-\frac{1}{2}\sigma^2(n-m)^2}
\rho_{nm}.
$$

This formula is not one deterministic random angle.  
It is the average effect of many shots where the rotation angle is random:

$$
\theta_\mathrm{shot} \sim \mathcal{N}(\theta,\sigma^2).
$$

Averaging over this Gaussian distribution gives

$$
\left\langle e^{i\theta_\mathrm{shot}(n-m)} \right\rangle
=
e^{i\theta(n-m)}
e^{-\frac{1}{2}\sigma^2(n-m)^2}.
$$

So:

- $\theta$ is the mean readout-induced rotation,
- $\sigma$ is the shot-to-shot uncertainty of that rotation,
- large $|n-m|$ coherences are dephased more strongly.

For a qubit initially in $|g\rangle$, we use

$$
\theta = \theta_g,
\qquad
\sigma = \sigma_g.
$$

For a qubit initially in $|e\rangle$, we include possible qubit decay during the readout.



#### 3. Excited-state decay during readout, assignment errors, and reset

If the qubit starts in $|e\rangle$, it can decay to $|g\rangle$ during the readout.  
We approximate this with a few hidden decay-time branches.

The no-jump probability is

$$
p_\mathrm{no\ jump}
=
e^{-T_\mathrm{ro}/T_1^\mathrm{qubit}}.
$$

For a jump branch where the qubit spends a fraction $f_i$ of the readout in $|e\rangle$, the cavity rotation is interpolated as

$$
\theta_i
=
f_i \theta_e + (1-f_i)\theta_g,
$$

and similarly for the phase diffusion,

$$
\sigma_i
=
f_i \sigma_e + (1-f_i)\sigma_g.
$$

The hidden branches are not sampled explicitly.  
They are summed into two unnormalized cavity states:

$$
\tilde{\rho}_\mathrm{rec\ g}
=
\text{all paths leading to recorded ground},
$$

$$
\tilde{\rho}_\mathrm{rec\ e}
=
\text{all paths leading to recorded excited}.
$$

Classical assignment errors are included with

$$
p_{g\to e}
=
P(\mathrm{record\ excited}\mid \mathrm{qubit\ ground}),
$$

$$
p_{e\to g}
=
P(\mathrm{record\ ground}\mid \mathrm{qubit\ excited}).
$$

Then the probabilities of the two recorded outcomes are

$$
P(\mathrm{rec\ g})
=
\mathrm{Tr}(\tilde{\rho}_\mathrm{rec\ g}),
$$

$$
P(\mathrm{rec\ e})
=
\mathrm{Tr}(\tilde{\rho}_\mathrm{rec\ e}).
$$

We sample only the recorded classical bit:

$$
m =
\begin{cases}
+1, & \mathrm{recorded\ ground}, \\
-1, & \mathrm{recorded\ excited}.
\end{cases}
$$

The conditional cavity state is

$$
\rho_\mathrm{cav}^{(m)}
=
\frac{\tilde{\rho}_m}{\mathrm{Tr}(\tilde{\rho}_m)}.
$$

Finally, after the readout, the qubit is reset to $|g\rangle$, so the output state is

$$
\rho_\mathrm{out}
=
|g\rangle\langle g|
\otimes
\rho_\mathrm{cav}^{(m)}.
$$

In short, the readout model does:

$$
\rho
\rightarrow
\text{qubit projection}
\rightarrow
\text{cavity loss + readout rotation + phase diffusion}
\rightarrow
\text{assignment-error recorded bit}
\rightarrow
|g\rangle\langle g| \otimes \rho_\mathrm{cav}^{(m)}.
$$

---

### 2. Idle and RNN feedback control

After the readout and reset, the state idles for the feedback latency:

$$
\rho_k^{(\mathrm{idle},m_k)}
=
e^{t_\mathrm{idle}\mathcal L_\mathrm{idle}}
\left[
\rho_k^{(\mathrm{ro},m_k)}
\right].
$$

The idle Liouvillian contains qubit relaxation, qubit dephasing, and cavity decay:

$$
\mathcal L_\mathrm{idle}\rho
=
\frac{1}{T_1^q}\mathcal D[\hat\sigma_-]\rho
+
\frac{1}{2T_\phi^q}\mathcal D[\hat\sigma_z]\rho
+
\frac{1}{T_1^a}\mathcal D[\hat a]\rho,
$$

with

$$
\frac{1}{T_\phi^q}
=
\frac{1}{T_2^q}
-
\frac{1}{2T_1^q}.
$$

The RNN receives the recorded readout bit and outputs the feedback-GRAPE controls:

$$
h_{k+1},u_k
=
\pi_\Theta(h_k,m_k),
$$

$$
u_k
\rightarrow
\{\epsilon_q(t_\ell),\epsilon_a(t_\ell)\}_{\ell=1}^N.
$$

During the feedback pulse, the Hamiltonian is piecewise constant:

$$
\hat H_\ell
=
\hat H_0+\hat H_{\mathrm{ctrl},\ell}.
$$

The drift Hamiltonian is

$$
\hat H_0
=
\chi\hat a^\dagger\hat a
\otimes
|e\rangle\langle e|
+
\frac{K_a}{2}
\hat a^{\dagger 2}\hat a^2.
$$

The control Hamiltonian is

$$
\hat H_{\mathrm{ctrl},\ell}
=
\mu_q\epsilon_{q,\ell}\hat\sigma_+
+
\mu_a i\epsilon_{a,\ell}^*\hat a^\dagger
+
\mathrm{h.c.}
$$

For each time interval,

$$
\hat U_\ell
=
e^{-i\Delta t_\ell \hat H_\ell},
\qquad
\mathcal U_\ell[\rho]
=
\hat U_\ell\rho\hat U_\ell^\dagger.
$$

---

### 3. Decoherence during control and training fidelity

The feedback-GRAPE pulse is not modeled as perfectly unitary.  
To approximate decoherence during the pulse, the control is split into $C$ chunks.

For each chunk, we define a half-idle channel

$$
\mathcal I_{1/2}
=
\exp
\left[
\frac{\delta t_\mathrm{chunk}}{2}
\mathcal L_\mathrm{idle}
\right],
\qquad
\delta t_\mathrm{chunk}
=
\frac{t_\mathrm{ctrl}}{C}.
$$

If $\mathcal U_c$ is the product of the unitary control intervals inside chunk $c$, the full control channel is approximated by

$$
\mathcal C_{\pi_\Theta(m_k)}
=
\prod_{c=1}^{C}
\left(
\mathcal I_{1/2}
\circ
\mathcal U_c
\circ
\mathcal I_{1/2}
\right).
$$

The final state after one feedback step is

$$
\rho_{k+1}
=
\mathcal C_{\pi_\Theta(m_k)}
\left[
\rho_k^{(\mathrm{idle},m_k)}
\right].
$$

The training fidelity is evaluated after the full feedback step:

$$
F_{k+1}
=
\langle\psi_\mathrm{target}|
\mathrm{Tr}_q[\rho_{k+1}]
|\psi_\mathrm{target}\rangle.
$$

The fidelity immediately after the readout can be logged as a diagnostic, but the reward uses the post-control state $\rho_{k+1}$.

## Trajectory

```python
def make_trajectory(N_cav, chi, t1_qub, t2_qub, t1_cav, t_idle, psi_target_cav):
    time_step = make_time_step(N_cav, chi, t1_qub, t2_qub, t1_cav, t_idle, psi_target_cav)
    
    def trajectory(key, flax_state, params, num_time_steps):
        
        def f(scan_carry, _):
            key, rho, carry = scan_carry
            key, subkey = jax.random.split(key)
            rho, carry, fid, meas_out, prob = time_step(subkey, rho, flax_state, params, carry)
            y = fid, meas_out, prob
            scan_carry = key, rho, carry
            return scan_carry, y
        
        key1, key2 = jax.random.split(key)
        carry = init_carry(key1, rec_features, inp_shape)
        psi_init = tensor(basis(2), psi_target_cav)
        scan_carry = key2, psi_init @ hconj(psi_init), carry
        _, ys = jax.lax.scan(f, scan_carry, length=num_time_steps)
        return ys
    
    return trajectory
```

## Trajectory with reward gradients

```python
def make_trajectory_with_rewgrad(N_cav, chi, t1_qub, t2_qub, t1_cav, t_idle, psi_target_cav):
    
    trajectory = make_trajectory(N_cav, chi, t1_qub, t2_qub, t1_cav, t_idle, psi_target_cav)
    
    def trajectory_with_rewgrad(key, flax_state, num_time_steps):
    
        def func(params):
            traj_out = trajectory(key, flax_state, params, num_time_steps)
            fid, meas_out, prob = traj_out
            # rew = -jnp.log(1 - fid.mean())
            # log_prob = jnp.log(prob).sum()
            # return rew + jax.lax.stop_gradient(rew) * log_prob, traj_out
    
            eps = 1e-8

            fid_mean = jnp.real(fid.mean())
            fid_mean = jnp.clip(fid_mean, 0.0, 1.0 - eps)

            rew = -jnp.log(1.0 - fid_mean)

            prob_safe = jnp.clip(prob, eps, 1.0)
            log_prob = jnp.log(prob_safe).sum()

            return rew + jax.lax.stop_gradient(rew) * log_prob, traj_out


        grad_func = jax.grad(func, has_aux=True)
        rewgrad, traj_out = grad_func(flax_state.params)
        return rewgrad, traj_out
    
    return trajectory_with_rewgrad
```

## Train on batch

```python
def make_train_on_batch(
        N_cav, chi,
        t1_qub, t2_qub, t1_cav, t_idle,
        psi_target_cav,
        batch_size, num_time_steps
    ):
    
    trajectory_with_rewgrad = make_trajectory_with_rewgrad(
        N_cav, chi, t1_qub, t2_qub, t1_cav, t_idle, psi_target_cav
    )
    
    trajectory_with_rewgrad_vec = jax.vmap(
            trajectory_with_rewgrad, in_axes=(0, None, None)
        )
    
    @jax.jit
    def train_on_batch(key, flax_state):
        keys = jax.random.split(key, batch_size)
        rewgrad, traj_out = trajectory_with_rewgrad_vec(keys, flax_state, num_time_steps)
        lossgrad = jax.tree.map(lambda x: -x.mean(axis=0), rewgrad)
        flax_state = flax_state.apply_gradients(grads=lossgrad) 
        return flax_state, traj_out
    
    return train_on_batch
```

## Test on batch

```python
def make_test_on_batch(
        N_cav, chi,
        t1_qub, t2_qub, t1_cav, t_idle,
        psi_target_cav,
        batch_size, num_time_steps
    ):
    
    trajectory = make_trajectory(
        N_cav, chi, t1_qub, t2_qub, t1_cav, t_idle, psi_target_cav
    )
    
    trajectory_vec = jax.vmap(
            trajectory, in_axes=(0, None, None, None)
        )
    
    @jax.jit
    def test_on_batch(key, flax_state):
        keys = jax.random.split(key, batch_size)
        traj_out = trajectory_vec(keys, flax_state, flax_state.params, num_time_steps)
        return traj_out
    
    return test_on_batch
```

# Normal GRAPE optimization with same Hamiltonian and target state

```python
from tqdm.auto import tqdm
from datetime import datetime
import os
from pathlib import Path

# Normal GRAPE: prepare |g,0> -> |g> tensor psi_target_cav
# Uses the already-defined feedback-GRAPE Hamiltonian through build_drive_htree.

psi_init_grape = tensor(basis(2), basis(N_cav, 0))       # |g,0>
psi_target_grape = tensor(basis(2), psi_target_cav)      # |g> ⊗ target cavity state

grape_optimizer = optax.adam(learning_rate=2e-3)


def grape_fidelity(psi, psi_target):
    overlap = hconj(psi_target) @ psi
    return jnp.abs(overlap[0, 0]) ** 2


def apply_grape_ctrl(grape_ctrl_coeffs):
    e_qub, e_cav = ctrl_from_coeffs(grape_ctrl_coeffs, bsplns_mids)

    # Uses the exact same drift + control Hamiltonians as feedback-GRAPE.
    htree = build_drive_htree(N_cav, chi, e_qub, e_cav)

    psi_t = sesolve_htree(htree, psi_init_grape, t_drive_intvs)
    return psi_t


@jax.jit
def grape_train_step(grape_ctrl_coeffs, grape_opt_state):
    def loss_fn(c):
        psi_t = apply_grape_ctrl(c)
        fid = grape_fidelity(psi_t[-1], psi_target_grape)
        return 1.0 - fid, fid

    (loss_val, fid_val), grads = jax.value_and_grad(loss_fn, has_aux=True)(grape_ctrl_coeffs)
    updates, grape_opt_state = grape_optimizer.update(grads, grape_opt_state)
    grape_ctrl_coeffs = optax.apply_updates(grape_ctrl_coeffs, updates)

    return grape_ctrl_coeffs, grape_opt_state, loss_val, fid_val
```

```python
# ============================================================
# Non-unitary GRAPE helpers
# Same method as the older notebook:
#   decay_fidelity(c) = fidelity after mesolve_htree(htree, c_ops, rho0, time_intervals)
# but using the new notebook Hamiltonian build_drive_htree(...)
# ============================================================

rho_init_grape = psi_init_grape @ hconj(psi_init_grape)

grape_decay_c_ops = build_c_ops(
    N_cav,
    t1_qub,
    t2_qub,
    t1_cav,
)


def decay_grape_fidelity(grape_ctrl_coeffs):
    e_qub, e_cav = ctrl_from_coeffs(grape_ctrl_coeffs, bsplns_mids)

    # Important: this is the Hamiltonian from the new notebook
    htree = build_drive_htree(N_cav, chi, e_qub, e_cav)

    rho_out = mesolve_htree(
        htree,
        grape_decay_c_ops,
        rho_init_grape,
        t_drive_intvs,
    )

    # depending on toolbox version, mesolve_htree may return final rho or trajectory
    rho_fin = rho_out[-1] if rho_out.ndim == 3 else rho_out

    fid = hconj(psi_target_grape) @ rho_fin @ psi_target_grape
    return jnp.real(fid[0, 0])
```

```python
# ============================================================
# Non-unitary GRAPE fine-tuning
# Rerunning this cell continues from the current decay-optimized pulse.
# ============================================================

n_steps_decay = 1000
learning_rate = 2e-7
decay_param_clip = 2.0

decay_optimizer = optax.adam(learning_rate=learning_rate)

try:
    grape_ctrl_coeffs_decay
    decay_opt_state
    grape_decay_fid_hist
    print("Continuing existing non-unitary GRAPE optimization.")
except NameError:
    grape_ctrl_coeffs_decay = jnp.clip(grape_ctrl_coeffs, -decay_param_clip, decay_param_clip)
    decay_opt_state = decay_optimizer.init(grape_ctrl_coeffs_decay)
    grape_decay_fid_hist = []
    print("Starting non-unitary GRAPE from current unitary-GRAPE pulse.")


@jax.jit
def decay_grape_train_step(ctrl_coeffs, opt_state):
    def loss_fn(c):
        fid = decay_grape_fidelity(c)
        return 1.0 - fid, fid

    (loss_value, fid_value), grads = jax.value_and_grad(loss_fn, has_aux=True)(ctrl_coeffs)

    updates, opt_state = decay_optimizer.update(grads, opt_state)
    ctrl_coeffs = optax.apply_updates(ctrl_coeffs, updates)

    # Same idea as the older notebook: keep coefficients bounded.
    ctrl_coeffs = jnp.clip(ctrl_coeffs, -decay_param_clip, decay_param_clip)

    return ctrl_coeffs, opt_state, loss_value, fid_value


print("Initial/current decay fidelity:", float(decay_grape_fidelity(grape_ctrl_coeffs_decay)))

pbar = tqdm(range(n_steps_decay), desc="non-unitary GRAPE")
for _ in pbar:
    grape_ctrl_coeffs_decay, decay_opt_state, decay_loss_val, decay_fid_val = decay_grape_train_step(
        grape_ctrl_coeffs_decay,
        decay_opt_state,
    )

    grape_decay_fid_hist.append(float(decay_fid_val))
    pbar.set_postfix(fid=f"{float(decay_fid_val):.8f}")

# Make downstream cells use the non-unitary-optimized pulse.
grape_ctrl_coeffs = grape_ctrl_coeffs_decay
grape_ctrl_coeffs_np = np.asarray(grape_ctrl_coeffs, dtype=float)

# Keep your existing export cell compatible if it uses grape_fid_hist[-1].
grape_fid_hist.append(float(grape_decay_fid_hist[-1]))

print("Final non-unitary-GRAPE fidelity:", grape_decay_fid_hist[-1])
print("grape_ctrl_coeffs shape:", grape_ctrl_coeffs_np.shape)

plt.figure(figsize=(6, 3))
plt.semilogy(1.0 - np.asarray(grape_decay_fid_hist))
plt.xlabel("non-unitary GRAPE step")
plt.ylabel("infidelity with decoherence")
plt.title("Non-unitary GRAPE fine-tuning")
plt.tight_layout()
plt.show()
```

```python
N_cav_dq = 40
N_lev_dq = 3

# Operators
Iq = dq.eye(N_lev_dq)
Ic = dq.eye(N_cav_dq)
a_q = dq.tensor(dq.destroy(N_lev_dq), Ic)
a_c = dq.tensor(Iq, dq.destroy(N_cav_dq))
a_q_dag = dq.dag(a_q)
a_c_dag = dq.dag(a_c)
n_q = a_q_dag @ a_q
n_phot = a_c_dag @ a_c

# Drift Hamiltonian
chi_mhz = -2 * jnp.pi * cfg["chi_kHz"] * 1e-3
aham_mhz = 2 * jnp.pi * cfg["anharmonicity_MHz"]
cavity_self_kerr = 2*jnp.pi*cfg["self_Kerr_kHz"]*1e-3  # cavity self-Kerr MHz

H0 = chi_mhz * n_phot @ n_q + (aham_mhz / 2) * a_q_dag @ a_q_dag @ a_q @ a_q + 0.5 * cavity_self_kerr * (a_c_dag @ a_c_dag @ a_c @ a_c)

# Control Hamiltonians (Matches the I+iQ derivation)
H_ctrl_qub_I = a_q + a_q_dag
H_ctrl_qub_Q = -1j * (a_q_dag - a_q)
H_ctrl_cav_I = 1j * (a_c_dag - a_c)
H_ctrl_cav_Q = (a_c + a_c_dag)

# Dissipation Rates (Units: MHz)
kappa_s = 1 / cfg["storage_T1_us"] 
kappa_t1 = 1 / cfg["qubit_T1_us"] 
kappa_t2 = 1 / cfg["qubit_T2_us"] 
kappa_phi = -kappa_t1 / 2 + kappa_t2

n_th_cav = cfg["storage_thermal_population"]
n_th_qubit = cfg["qubit_thermal_population"]

rate_down_cav = kappa_s * (1 - n_th_cav)
rate_up_cav = kappa_s * n_th_cav
rate_down_qubit = kappa_t1 * (1 - n_th_qubit)
rate_up_qubit = kappa_t1 * n_th_qubit
rate_phase = kappa_phi

# Jump Operators
l_down_cav = jnp.sqrt(rate_down_cav) * a_c
l_up_cav = jnp.sqrt(rate_up_cav) * a_c_dag
l_down_qubit = jnp.sqrt(rate_down_qubit) * a_q
l_up_qubit = jnp.sqrt(rate_up_qubit) * a_q_dag
l_phase = jnp.sqrt(2 * rate_phase) * n_q

jump_ops = [l_down_cav, l_up_cav, l_down_qubit, l_up_qubit, l_phase]
```

```python
# Initial and Target States
psi_init_dq = dq.tensor(dq.basis(N_lev_dq, 0), dq.basis(N_cav_dq, 0))
rho0 = dq.todm(psi_init_dq)




psi_target_cav_dq = dq.coherent(N_cav_dq, 2.0)-dq.coherent(N_cav_dq, -2.0)
# psi_target_cav_dq =dq.basis(N_cav_dq,2)
psi_target_cav_dq = psi_target_cav_dq/dq.norm(psi_target_cav_dq)

psi_target_dq = dq.tensor(dq.basis(N_lev_dq, 0),psi_target_cav_dq)
rho_target = dq.todm(psi_target_dq)

# Construct Time-Dependent Hamiltonian
t_edges = jnp.asarray(t_edges_us)
eps_qub_I_jnp = jnp.asarray(eps_qubit_I)
eps_qub_Q_jnp = jnp.asarray(eps_qubit_Q)
eps_cav_I_jnp = jnp.asarray(eps_cav_I)
eps_cav_Q_jnp = jnp.asarray(eps_cav_Q)

H_qub_I_t = dq.pwc(t_edges, eps_qub_I_jnp, H_ctrl_qub_I)
H_qub_Q_t = dq.pwc(t_edges, eps_qub_Q_jnp, H_ctrl_qub_Q)
H_cav_I_t = dq.pwc(t_edges, eps_cav_I_jnp, H_ctrl_cav_I)
H_cav_Q_t = dq.pwc(t_edges, eps_cav_Q_jnp, H_ctrl_cav_Q)

H_t = H0 + H_qub_I_t + H_qub_Q_t + H_cav_I_t + H_cav_Q_t

# Integration settings
tsave = t_edges # Save at every 1 ns edge
method = dq.method.Tsit5(rtol=1e-8, atol=1e-10, max_steps=1_000_000)
opts   = dq.Options(save_states=True, progress_meter=True)

# Solve
result_dq = dq.mesolve(
    H=H_t,
    jump_ops=jump_ops,
    rho0=rho0,
    tsave=tsave,
    exp_ops=[n_phot, n_q], 
    method=method,
    options=opts,
)

rho_final = result_dq.states[-1]
F_final = float(dq.fidelity(rho_final, rho_target))
print(f"Fidelity (mesolve, with decoherence): {F_final:.8f}")
```

```python
import jax
import jax.numpy as jnp
import jax.scipy.linalg as jsp_linalg

def measure_sz(
    key,
    rho_init,
    chi=chi,
    t_feedback=0.42,
    readout_duration=1.6,
    theta_g_deg=-18.67,
    theta_e_deg=+13.78,
):
    """
    Projective qubit measurement + experimentally calibrated readout-induced
    cavity rotation.

    Convention:
        outcome_indx = 0  -> qubit ground    -> outcome = +1
        outcome_indx = 1  -> qubit excited   -> outcome = -1

    Readout-induced cavity rotations:
        qubit ground  : theta_g_deg = -18.67 deg, clockwise
        qubit excited : theta_e_deg = +13.78 deg, anticlockwise

    Positive theta means anticlockwise rotation in phase space.
    """

    # Measurement probabilities: [p_ground, p_excited]
    probs = rho_init.diagonal().reshape(2, -1).sum(axis=1).real

    outcome_indx = jax.random.choice(key, 2, p=probs)
    prob = probs[outcome_indx]

    N_cav = rho_init.shape[-1] // 2

    # Select the cavity block corresponding to the measurement outcome
    rho_cav = jnp.where(
        outcome_indx == 0,
        rho_init[:N_cav, :N_cav],
        rho_init[N_cav:, N_cav:],
    )

    # ------------------------------------------------------------
    # Readout-induced cavity rotation, conditioned on qubit outcome
    # ------------------------------------------------------------
    theta_deg = jnp.where(
        outcome_indx == 0,
        theta_g_deg,
        theta_e_deg,
    )

    theta = theta_deg * jnp.pi / 180.0

    a_c = destroy(N_cav)
    adag_c = hconj(a_c)
    n_c = adag_c @ a_c

    # Positive theta gives anticlockwise Wigner rotation.
    U_ro = jsp_linalg.expm(1j * theta * n_c)

    rho_cav = U_ro @ rho_cav @ hconj(U_ro)

    # Qubit reset to ground
    rho_fin_qub = basis(2) @ hconj(basis(2))
    rho_fin = tensor(rho_fin_qub, rho_cav)

    # Renormalize after projection
    rho_fin = rho_fin / jnp.trace(rho_fin)

    # Outcome definition:
    # ground -> +1, excited -> -1
    outcome = 1 - 2 * outcome_indx

    return outcome, prob, rho_fin
```
