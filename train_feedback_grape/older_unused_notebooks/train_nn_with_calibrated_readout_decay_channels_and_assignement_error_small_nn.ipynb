## Driven Hamiltonian

````python
def build_drive_htree(N_cav, chi, e_qub, e_cav):
    a = tensor(identity(2), destroy(N_cav))
    adag = hconj(a)
    n_phot = adag @ a
    sigz = tensor(sigma.z, identity(N_cav))
    sigp = tensor(sigma.p, identity(N_cav))
    one = tensor(identity(2), identity(N_cav))

    H_disp = (chi / 2) * n_phot @ (one - sigz)
    H_kerr = 0.5 * K_c * adag @ adag @ a @ a

    # Combined drift Hamiltonian
    H_drift = H_disp + H_kerr
    
    ht = [
        [H_drift, 1, 1, 0],
        [mu_qub * sigp, e_qub, 1, 1],                
        [mu_cav * adag, 1j *jnp.conj(e_cav), 1, 1] 
    ]
    return ht

def get_uevol_matrix(N_cav, chi, e_qub, e_cav, dt):
    htree = build_drive_htree(N_cav, chi, e_qub, e_cav)
    hmat = hmat_from_htree(htree)
    return evol_hdt_exp(hmat, dt)

def make_drive(N_cav, chi):
    
    def drive(rho_init, es_qub, es_cav, dts):
        
        def f(rho, x):
            e_qub, e_cav, dt = x
            uevol = get_uevol_matrix(N_cav, chi, e_qub, e_cav, dt)[0]
            return uevol @ rho @ hconj(uevol), None
    
        rho_fin, _ = jax.lax.scan(f, rho_init, [es_qub, es_cav, dts])
        return rho_fin
    
    return drive

````

## Relaxation

````python
import jax.scipy.linalg as jsp_linalg


def qubit_ops():
    g = basis(2)       # |g> = index 0
    e = basis(2, 1)    # |e> = index 1

    Pg = g @ hconj(g)
    Pe = e @ hconj(e)

    sm = g @ hconj(e)  # |g><e|
    sp = e @ hconj(g)  # |e><g|

    sz = Pg - Pe       # diag(+1, -1)

    return g, e, Pg, Pe, sm, sp, sz


def liouv_relax_channel(c):
    """
    Dissipator D[c] in row-major vectorization convention.

    For row-major vectorization:
        vec(A rho B) = (A kron B.T) vec(rho)

    Therefore:
        c rho c^\dagger -> c kron c.conj()
    """
    D = len(c)

    l = tensor(c, c.conj())
    l -= 0.5 * tensor(hconj(c) @ c, identity(D))
    l -= 0.5 * tensor(identity(D), (hconj(c) @ c).T)

    return l


def liouv_hamiltonian(H):
    """
    Hamiltonian Liouvillian for -i[H, rho],
    using the same row-major vectorization convention.
    """
    D = len(H)
    I = identity(D)

    return -1j * (
        tensor(H, I)
        -
        tensor(I, H.T)
    )


def build_c_ops(N_cav, t1_qub, t2_qub, t1_cav):
    """
    Collapse operators for the full qubit+cavity system.

    Units:
        all times are in us.
    """
    get_t_phi_qub = lambda t1, t2: 1.0 / ((1.0 / t2) - (1.0 / t1) / 2.0)
    t_phi_qub = get_t_phi_qub(t1_qub, t2_qub)

    g, e, Pg, Pe, sm, sp, sz = qubit_ops()

    sig_down = tensor(sm, identity(N_cav))
    sigz = tensor(sz, identity(N_cav))
    a = tensor(identity(2), destroy(N_cav))

    c_ops = [
        (1.0 / jnp.sqrt(t1_qub)) * sig_down,
        (1.0 / jnp.sqrt(t_phi_qub)) * sigz / jnp.sqrt(2.0),
        (1.0 / jnp.sqrt(t1_cav)) * a,
    ]

    return c_ops


def build_idle_hamiltonian(N_cav, chi):
    """
    Drift Hamiltonian during a free idle.

    This matches the convention of build_drive_htree:

        H_disp = (chi / 2) * n * (1 - sigma_z)

    With sigma_z = |g><g| - |e><e|, this is equivalent to

        H_disp = chi * n * |e><e|.

    Therefore the cavity rotates under chi only when the qubit is in |e>.
    """
    g, e, Pg, Pe, sm, sp, sz = qubit_ops()

    a = tensor(identity(2), destroy(N_cav))
    adag = hconj(a)
    n_phot = adag @ a

    Pe_full = tensor(Pe, identity(N_cav))

    H_disp = chi * n_phot @ Pe_full
    H_kerr = 0.5 * K_c * adag @ adag @ a @ a

    return H_disp + H_kerr


def eval_idling_matrix(cs, t):
    """
    Dissipative-only idle matrix.

    Use this inside the feedback-GRAPE control pulse splitting,
    because the unitary control chunks already include H_disp and H_kerr.
    """
    L = sum(liouv_relax_channel(c) for c in cs)
    return jsp_linalg.expm(L * t)


def eval_idling_matrix_with_drift(N_cav, chi, cs, t):
    """
    Full idle matrix including:
        - H_disp,
        - H_kerr,
        - qubit T1,
        - qubit dephasing,
        - cavity T1.
    """
    H_idle = build_idle_hamiltonian(N_cav, chi)

    L = liouv_hamiltonian(H_idle)

    for c in cs:
        L = L + liouv_relax_channel(c)

    return jsp_linalg.expm(L * t)


def make_idle(N_cav, t1_qub, t2_qub, t1_cav, t):
    """
    Dissipative-only idle.

    This includes:
        - qubit T1,
        - qubit dephasing,
        - cavity T1.

    It does NOT include chi or Kerr.

    Use this:
        - inside make_drive_with_interleaved_decay,
        - inside the 400 ns reset-delay model when the fitted readout angles
          already include the 400 ns rotation.
    """
    cs = build_c_ops(N_cav, t1_qub, t2_qub, t1_cav)
    idling_matrix = eval_idling_matrix(cs, t)

    def idle(rho):
        rho_liouv = rho.reshape(-1, 1)
        rho_liouv = idling_matrix @ rho_liouv
        return rho_liouv.reshape(rho.shape)

    return idle


def make_idle_with_drift(N_cav, chi, t1_qub, t2_qub, t1_cav, t):
    """
    Physical free idle.

    This includes:
        - H_disp,
        - H_kerr,
        - qubit T1,
        - qubit dephasing,
        - cavity T1.

    Use this after the reset, because if the reset fails and leaves the qubit
    in |e>, the cavity should rotate under chi.
    """
    cs = build_c_ops(N_cav, t1_qub, t2_qub, t1_cav)
    idling_matrix = eval_idling_matrix_with_drift(N_cav, chi, cs, t)

    def idle(rho):
        rho_liouv = rho.reshape(-1, 1)
        rho_liouv = idling_matrix @ rho_liouv
        return rho_liouv.reshape(rho.shape)

    return idle


def qubit_excited_population(rho):
    """
    Diagnostic:
        P_e = Tr[ <e|rho|e> ].
    """
    N_cav = rho.shape[-1] // 2
    return jnp.real(jnp.trace(rho[N_cav:, N_cav:]))


print("sigma.m =", sigma.m)
````

# Control with interleaved decay

````python
def make_drive_with_interleaved_decay(
    N_cav,
    chi,
    t1_qub,
    t2_qub,
    t1_cav,
    t_control=1.408,      # us
    n_decay_chunks=4,     # start with 4 or 8
):
    """
    Approximate dissipative evolution during the feedback-GRAPE pulse.

    Instead of:
        full unitary drive, then full idle

    use Strang splitting:
        half idle -> unitary chunk -> half idle
        repeated over chunks.

    This is a better approximation when the qubit can be populated during
    the control pulse.
    """

    drive = make_drive(N_cav, chi)

    n_intvs = len(t_drive_intvs)

    # Fixed Python slices, so JAX sees a static unrolled loop.
    edges = np.linspace(0, n_intvs, n_decay_chunks + 1, dtype=int)
    chunk_slices = [(int(edges[i]), int(edges[i + 1])) for i in range(n_decay_chunks)]

    dt_chunk = t_control / n_decay_chunks

    # Half-step dissipative evolution for Strang splitting.
    idle_half_chunk = make_idle(
        N_cav,
        t1_qub,
        t2_qub,
        t1_cav,
        0.5 * dt_chunk,
    )

    def drive_with_decay(rho, es_qub, es_cav):
        for start, stop in chunk_slices:
            if stop > start:
                # Half dissipative step before this unitary chunk
                rho = idle_half_chunk(rho)

                # Unitary control chunk
                rho = drive(
                    rho,
                    es_qub[start:stop],
                    es_cav[start:stop],
                    t_drive_intvs[start:stop],
                )

                # Half dissipative step after this unitary chunk
                rho = idle_half_chunk(rho)

        return rho

    return drive_with_decay
````

## Effective readout with delayed conditional reset

The readout model should not reset the qubit immediately after sampling the recorded bit.

The important physical sequence is now:

$$
\rho
\rightarrow
\text{qubit projection}
\rightarrow
\text{readout-induced cavity backaction}
\rightarrow
\text{hidden qubit state after readout}
\rightarrow
400\ \mathrm{ns}\ \text{reset delay}
\rightarrow
\text{conditional } \pi \text{ pulse if recorded excited}.
$$

This matters because the following bad path can occur:

$$
\mathrm{recorded}\ e
\rightarrow
e \to g \text{ during the 400 ns delay}
\rightarrow
\pi
\rightarrow
g \to e.
$$

So the network is not punished by an artificial cost.  
It is punished because the simulated physics can leave the qubit in the wrong state.

Important convention:

- recorded ground gives `meas_out = +1`,
- recorded excited gives `meas_out = -1`.

The RNN still receives only this recorded classical bit.

### Important calibration point

The fitted angles `theta_g_deg` and `theta_e_deg` were measured experimentally using the full sequence

$$
\text{readout}
\rightarrow
400\ \mathrm{ns}\ \text{delay}
\rightarrow
\text{conditional } \pi
\rightarrow
\text{Wigner}.
$$

Therefore they already include the deterministic cavity rotation during the 400 ns delay.

Because of this, inside the readout model, the explicit 400 ns reset-delay channel is dissipative-only:

```python
reset_delay_channel = make_idle(...)
```

not

```python
reset_delay_channel = make_idle_with_drift(...)
```

Otherwise the cavity rotation during the 400 ns delay would be counted twice.

After the reset, however, the normal idle should still use `make_idle_with_drift`, because if the reset leaves the qubit in `|e>`, the cavity must rotate under the dispersive Hamiltonian.

````python
import jax
import jax.numpy as jnp
import jax.scipy.linalg as jsp_linalg


# ============================================================
# Effective cavity-only channels used inside the readout model
# ============================================================

def make_cavity_loss_only(N_cav, t1_cav, t):
    """
    Cavity amplitude damping during time t.

    Units:
        t and t1_cav must be in the same units, here us.

    This uses the same Liouville convention as the rest of the notebook.
    """
    a = destroy(N_cav)
    c = (1.0 / jnp.sqrt(t1_cav)) * a

    L = liouv_relax_channel(c)
    M = jsp_linalg.expm(L * t)

    def cavity_loss(rho_cav):
        rho_vec = rho_cav.reshape(-1, 1)
        rho_vec = M @ rho_vec
        return rho_vec.reshape(rho_cav.shape)

    return cavity_loss


def apply_phase_rotation_and_diffusion(rho_cav, theta_rad, sigma_theta_rad):
    """
    Averaged random phase-space rotation.

    Deterministic part:
        rho_nm -> exp(i theta (n-m)) rho_nm

    Gaussian shot-to-shot phase uncertainty:
        rho_nm -> exp[-0.5 sigma_theta^2 (n-m)^2] rho_nm
    """
    N = rho_cav.shape[0]
    n = jnp.arange(N)

    dn = n[:, None] - n[None, :]

    phase = jnp.exp(1j * theta_rad * dn)
    deph = jnp.exp(-0.5 * (sigma_theta_rad ** 2) * (dn ** 2))

    return rho_cav * phase * deph


# ============================================================
# Effective readout model with delayed conditional reset
# ============================================================

def make_effective_readout(
    N_cav,
    chi,
    t1_qub,
    t2_qub,
    t1_cav,
    psi_target_cav,
    *,
    readout_duration,       # us, physical readout pulse duration
    reset_delay,            # us, delay before the conditional pi reset

    # Mean readout-induced cavity rotations.
    # Important:
    # In your current calibration, these already include:
    #     readout -> 400 ns delay -> conditional pi -> Wigner.
    theta_g_deg,
    theta_e_deg,

    # Readout-induced phase diffusion
    sigma_g_deg,
    sigma_e_deg,

    # Readout assignment errors
    # p_g_to_e = P(record excited | qubit effectively ground)
    # p_e_to_g = P(record ground   | qubit effectively excited)
    p_g_to_e,
    p_e_to_g,

    # Number of effective decay-time bins for e -> g during readout
    n_jump_bins,
):
    """
    Effective measurement + readout backaction + delayed conditional reset.

    Output convention:
        recorded ground  -> meas_out = +1
        recorded excited -> meas_out = -1

    Physics included:
        1. Projective qubit measurement into rho_gg / rho_ee blocks.
        2. Cavity loss during the readout pulse.
        3. Experimentally calibrated readout-induced cavity rotation.
        4. Readout-induced phase diffusion.
        5. Qubit T1 decay during the readout using hidden jump-time bins.
        6. Classical readout assignment errors.
        7. Qubit T1/dephasing/cavity loss during the 400 ns reset delay.
        8. Conditional pi reset pulse only if the recorded bit is excited.

    Important calibration convention:
        theta_g_deg and theta_e_deg already include the deterministic cavity
        rotation of the full experimental sequence:

            readout -> 400 ns delay -> conditional pi -> Wigner

        Therefore reset_delay_channel must NOT include H_disp or H_kerr.
        It is dissipative-only, otherwise the 400 ns cavity rotation is
        double-counted.

    Important modeling convention:
        The sampled variable is still only the recorded bit.
        The hidden readout-decay branches are summed analytically.
    """

    cavity_loss_ro = make_cavity_loss_only(N_cav, t1_cav, readout_duration)

    # Important:
    # Use make_idle, not make_idle_with_drift.
    #
    # The fitted theta_g/theta_e already include the cavity rotation from the
    # 400 ns delay. Here we only want to allow the physical qubit to decay
    # before the conditional pi pulse.
    reset_delay_channel = make_idle(
        N_cav=N_cav,
        t1_qub=t1_qub,
        t2_qub=t2_qub,
        t1_cav=t1_cav,
        t=reset_delay,
    )

    g, e, Pg, Pe, sm, sp, sz = qubit_ops()
    X_pi = tensor(sm + sp, identity(N_cav))

    theta_g = jnp.deg2rad(theta_g_deg)
    theta_e = jnp.deg2rad(theta_e_deg)

    sigma_g = jnp.deg2rad(sigma_g_deg)
    sigma_e = jnp.deg2rad(sigma_e_deg)

    # ------------------------------------------------------------
    # Effective excited-state decay branches during readout
    # ------------------------------------------------------------
    #
    # Branch 0:
    #     no jump during the readout.
    #     The physical qubit is still excited after the readout.
    #
    # Branch 1..n_jump_bins:
    #     jump happens inside a time bin.
    #     The physical qubit is ground after the readout.
    #
    # If the jump happens after fraction f of the readout,
    # approximate the cavity rotation as
    #
    #     theta_branch = f theta_e + (1 - f) theta_g
    #
    # and similarly for the phase diffusion.
    # ------------------------------------------------------------

    edges = jnp.linspace(0.0, 1.0, n_jump_bins + 1)
    mids = 0.5 * (edges[:-1] + edges[1:])

    gammaT = readout_duration / t1_qub

    p_no_jump = jnp.exp(-gammaT)
    p_jump_bins = jnp.exp(-gammaT * edges[:-1]) - jnp.exp(-gammaT * edges[1:])

    frac_excited = jnp.concatenate([
        jnp.array([1.0]),  # no jump branch
        mids,
    ])

    e_branch_weights = jnp.concatenate([
        jnp.array([p_no_jump]),
        p_jump_bins,
    ])

    # Numerical safety.
    e_branch_weights = e_branch_weights / jnp.sum(e_branch_weights)

    e_branch_thetas = frac_excited * theta_e + (1.0 - frac_excited) * theta_g
    e_branch_sigmas = frac_excited * sigma_e + (1.0 - frac_excited) * sigma_g

    # Effective probability of recording excited for an excited-origin branch.
    # If f=0, the readout looks ground-like.
    # If f=1, the readout looks excited-like.
    p_rec_e_for_e_branch = (
        p_g_to_e
        + frac_excited * ((1.0 - p_e_to_g) - p_g_to_e)
    )
    p_rec_g_for_e_branch = 1.0 - p_rec_e_for_e_branch

    # Ground-origin branch assignment probabilities.
    p_rec_g_for_g_branch = 1.0 - p_g_to_e
    p_rec_e_for_g_branch = p_g_to_e

    def branch_channel(rho_cav_unnorm, theta, sigma):
        """
        Linear channel on an unnormalized cavity block.
        """
        rho = cavity_loss_ro(rho_cav_unnorm)
        rho = apply_phase_rotation_and_diffusion(rho, theta, sigma)
        return rho

    def make_block_diagonal_state(rho_g, rho_e):
        """
        Build a full qubit+cavity density matrix from two cavity blocks:

            rho = |g><g| tensor rho_g + |e><e| tensor rho_e.

        The input blocks may be unnormalized.
        """
        rho_full = jnp.zeros(
            (2 * N_cav, 2 * N_cav),
            dtype=rho_g.dtype,
        )

        rho_full = rho_full.at[:N_cav, :N_cav].set(rho_g)
        rho_full = rho_full.at[N_cav:, N_cav:].set(rho_e)

        return rho_full

    def readout(key, rho_init):
        """
        Effective readout step.

        Input:
            rho_init:
                full qubit+cavity density matrix,
                shape (2*N_cav, 2*N_cav)

        Output:
            meas_out:
                +1 for recorded ground, -1 for recorded excited

            prob:
                probability of the sampled recorded outcome

            rho_fin:
                full qubit+cavity density matrix after:
                    readout,
                    reset delay,
                    conditional pi reset

            fid_after_readout:
                diagnostic cavity fidelity after readout + delayed reset
        """

        eps = 1e-12

        # Qubit blocks. The qubit coherences are discarded by the measurement.
        rho_gg = rho_init[:N_cav, :N_cav]
        rho_ee = rho_init[N_cav:, N_cav:]

        # ========================================================
        # Ground-origin branch
        # ========================================================
        # If the qubit starts in ground, we assume it remains ground
        # during the readout. Only classical assignment error can make
        # the recorded bit excited.
        rho_g_branch = branch_channel(rho_gg, theta_g, sigma_g)

        rho_rec_g_true_g = p_rec_g_for_g_branch * rho_g_branch
        rho_rec_e_true_g = p_rec_e_for_g_branch * rho_g_branch

        rho_rec_g_true_e = jnp.zeros_like(rho_g_branch)
        rho_rec_e_true_e = jnp.zeros_like(rho_g_branch)

        # ========================================================
        # Excited-origin hidden branches
        # ========================================================
        for i in range(n_jump_bins + 1):
            rho_e_i = branch_channel(
                rho_ee,
                e_branch_thetas[i],
                e_branch_sigmas[i],
            )

            rho_e_i = e_branch_weights[i] * rho_e_i

            if i == 0:
                # No jump during readout.
                # The physical qubit is still excited after readout.
                rho_rec_g_true_e = rho_rec_g_true_e + p_rec_g_for_e_branch[i] * rho_e_i
                rho_rec_e_true_e = rho_rec_e_true_e + p_rec_e_for_e_branch[i] * rho_e_i
            else:
                # Jump happened during readout.
                # The physical qubit is ground after readout.
                rho_rec_g_true_g = rho_rec_g_true_g + p_rec_g_for_e_branch[i] * rho_e_i
                rho_rec_e_true_g = rho_rec_e_true_g + p_rec_e_for_e_branch[i] * rho_e_i

        # ========================================================
        # Recorded outcome probabilities
        # ========================================================
        rho_record_g = rho_rec_g_true_g + rho_rec_g_true_e
        rho_record_e = rho_rec_e_true_g + rho_rec_e_true_e

        p_g_raw = jnp.real(jnp.trace(rho_record_g))
        p_e_raw = jnp.real(jnp.trace(rho_record_e))

        p_raw = jnp.array([p_g_raw, p_e_raw])
        p_raw = jnp.maximum(p_raw, 0.0)

        p_sample = p_raw / (jnp.sum(p_raw) + eps)

        outcome_indx = jax.random.choice(key, 2, p=p_sample)
        prob = p_sample[outcome_indx]

        

        # ========================================================
        # Conditional physical qubit+cavity state after readout,
        # but before the reset delay
        # ========================================================

        rho_pre_delay_rec_g = make_block_diagonal_state(
            rho_rec_g_true_g / (p_g_raw + eps),
            rho_rec_g_true_e / (p_g_raw + eps),
        )

        rho_pre_delay_rec_e = make_block_diagonal_state(
            rho_rec_e_true_g / (p_e_raw + eps),
            rho_rec_e_true_e / (p_e_raw + eps),
        )

        rho_pre_delay = jnp.where(
            outcome_indx == 0,
            rho_pre_delay_rec_g,
            rho_pre_delay_rec_e,
        )

        rho_pre_delay = rho_pre_delay / (jnp.trace(rho_pre_delay) + eps)

        fid_after_readout = fidelity_cav(rho_pre_delay, psi_target_cav) #after readout but before readout backaction

        # ========================================================
        # Physical reset delay before conditional pi pulse
        # ========================================================
        # This is dissipative-only because theta_g/theta_e already include
        # the deterministic cavity rotation of:
        #
        #     readout -> 400 ns delay -> conditional pi -> Wigner
        #
        # But this delay still allows the important bad path:
        #
        #     recorded excited
        #     -> physical qubit decays e -> g during reset_delay
        #     -> conditional pi pulse flips g -> e
        #
        rho_after_delay = reset_delay_channel(rho_pre_delay)

        # Conditional pi pulse:
        # apply pi only if the recorded bit was excited.
        rho_after_pi = X_pi @ rho_after_delay @ hconj(X_pi)

        rho_fin = jnp.where(
            outcome_indx == 1,
            rho_after_pi,
            rho_after_delay,
        )

        rho_fin = rho_fin / (jnp.trace(rho_fin) + eps)

        

        # recorded ground -> +1
        # recorded excited -> -1
        meas_out = 1 - 2 * outcome_indx

        return meas_out, prob, rho_fin, fid_after_readout

    return readout
````

````python
def init_carry(key, rec_features, inp_shape):
    gru_cell = nn.GRUCell(features=rec_features)
    return gru_cell.initialize_carry(key, inp_shape)

def create_flax_state(key,
                      rec_features, dense_features,
                      inp_shape, learning_rate,
                      *, print_summary):
    
    class Model(nn.Module):
    
        @nn.compact
        def __call__(self, carry, x):
            
            gru_cell = nn.GRUCell(features=rec_features)
            carry, x = gru_cell(carry, x)
            
            for ndf in dense_features:
                x = nn.Dense(ndf)(x)
                x = jax.nn.relu(x)
                
            out_layer = nn.Dense(4 * num_bspln)
            x = out_layer(x)
            x = x.reshape(4, num_bspln)
            
            return carry, x

    model = Model()

    fake_key = jax.random.key(0)
    fake_carry = init_carry(fake_key, rec_features, inp_shape)
    fake_inp = jnp.ones(inp_shape)
    params = model.init(key, fake_carry, fake_inp)["params"]

    # optimizer = optax.adam(learning_rate=learning_rate)
    optimizer = optax.chain( #avoid unlikely trajectory to create catastrophic update
    optax.clip_by_global_norm(1.0),
    optax.adam(learning_rate=learning_rate),
    )

    flax_state = TrainState.create(apply_fn=model.apply, params=params, tx=optimizer)

    if print_summary:
        print(model.tabulate(fake_key, fake_carry, fake_inp))

    return flax_state

def apply_nn(flax_state, params, carry, inp):
    return flax_state.apply_fn({'params': params}, carry, inp)

````

## Time step

One feedback step is now modeled as:

1. effective readout pulse,
2. hidden readout decay branches,
3. physical delay before reset,
4. conditional $\pi$ pulse if the recorded bit is excited,
5. free idle including $\chi$ and decay,
6. RNN interrogation using only the recorded bit,
7. feedback-GRAPE control with interleaved dissipation.

The important change is that the reset is no longer perfect and instantaneous.  
The simulation now allows this physical error path:

$$
\mathrm{recorded}\ e
\rightarrow
e \to g \text{ during reset delay}
\rightarrow
\pi \text{ pulse}
\rightarrow
g \to e.
$$

Then, if the qubit is accidentally left excited, the following idle evolves with the dispersive Hamiltonian, so the cavity rotates conditionally on the wrong qubit state.

````python
def make_time_step(
    N_cav,
    chi,
    t1_qub,
    t2_qub,
    t1_cav,
    t_idle,
    psi_target_cav,
    t_control=1.408,       # us
    n_decay_chunks=4,
):
    readout = make_effective_readout(
        N_cav=N_cav,
        chi=chi,
        t1_qub=t1_qub,
        t2_qub=t2_qub,
        t1_cav=t1_cav,
        psi_target_cav=psi_target_cav,

        # Physical readout pulse duration only.
        # Do not use 1.6 + 0.4 here.
        #
        # The fitted theta_g/theta_e already include:
        #     readout -> 400 ns delay -> conditional pi -> Wigner
        readout_duration=1.6,

        # Physical delay before the conditional pi pulse.
        # Inside make_effective_readout this is dissipative-only,
        # so it does not double-count the fitted cavity rotation.
        reset_delay=0.42,

        theta_g_deg=-18.67,
        theta_e_deg=+13.78,

        sigma_g_deg=0.5,
        sigma_e_deg=2.0,

        p_g_to_e=0.00,
        p_e_to_g=0.04,

        n_jump_bins=5,
    )

    # After the reset, the normal idle must include chi.
    # If the reset accidentally leaves the qubit in |e>, the cavity rotates.
    idle = make_idle_with_drift(
        N_cav=N_cav,
        chi=chi,
        t1_qub=t1_qub,
        t2_qub=t2_qub,
        t1_cav=t1_cav,
        t=t_idle,
    )

    drive_with_decay = make_drive_with_interleaved_decay(
        N_cav=N_cav,
        chi=chi,
        t1_qub=t1_qub,
        t2_qub=t2_qub,
        t1_cav=t1_cav,
        t_control=t_control,
        n_decay_chunks=n_decay_chunks,
    )

    def time_step(key, rho, flax_state, params, carry):
        # 1. Measurement + calibrated readout backaction
        #    + reset-delay decay
        #    + conditional pi reset.
        meas_out, prob, rho, fid_after_readout = readout(key, rho)

        # 2. Physical idle / feedback latency after reset.
        #    This includes chi and Kerr.
        rho = idle(rho)

        # 3. RNN receives only the recorded bit.
        rnn_inp = jnp.array([meas_out], dtype=float)
        carry, ctrl_coeffs = apply_nn(flax_state, params, carry, rnn_inp)

        # 4. Apply feedback-GRAPE pulse with interleaved decoherence.
        es_qub, es_cav = ctrl_from_coeffs(ctrl_coeffs, bsplns_mids)
        rho = drive_with_decay(rho, es_qub, es_cav)

        # 5. Fidelity after the controller had a chance to correct.
        fid_after_control = fidelity_cav(rho, psi_target_cav)

        return rho, carry, fid_after_readout, fid_after_control, meas_out, prob

    return time_step
````

## Feedback-GRAPE simulation model with delayed reset

At each feedback step, the input state is the full qubit+cavity density matrix $\rho_k$.

The updated simulation step is

$$
\rho_k
\rightarrow
\text{effective readout}
\rightarrow
\text{hidden qubit state after readout}
\rightarrow
400\ \mathrm{ns}\ \text{reset delay}
\rightarrow
\text{conditional reset}
\rightarrow
\text{idle}
\rightarrow
\text{RNN control + dissipation}
\rightarrow
\rho_{k+1}.
$$

The recorded readout bit is

$$
m_k =
\begin{cases}
+1, & \mathrm{recorded\ ground},\\
-1, & \mathrm{recorded\ excited}.
\end{cases}
$$

Only this recorded bit $m_k$ is given to the RNN.

---

### 1. Qubit projection

In the basis $\{|g\rangle, |e\rangle\}$, the full density matrix is written as

$$
\rho =
\begin{pmatrix}
\rho_{gg} & \rho_{ge} \\
\rho_{eg} & \rho_{ee}
\end{pmatrix}.
$$

The readout destroys the qubit coherences, so the model keeps only

$$
\rho_{gg}, \qquad \rho_{ee}.
$$

Their traces are

$$
p_g = \mathrm{Tr}(\rho_{gg}), 
\qquad
p_e = \mathrm{Tr}(\rho_{ee}).
$$

---

### 2. Effective cavity backaction during readout

For a cavity block $\rho_\mathrm{cav}$, the readout applies

$$
\rho_\mathrm{cav}
\mapsto
\mathcal R_{\theta,\sigma}
\left[
\mathcal L_{T_\mathrm{ro}}(\rho_\mathrm{cav})
\right].
$$

Here $\mathcal L_{T_\mathrm{ro}}$ is cavity loss during the readout pulse.

The phase-space rotation plus shot-to-shot phase diffusion is applied in the Fock basis as

$$
\left[\mathcal R_{\theta,\sigma}(\rho)\right]_{nm}
=
e^{i\theta(n-m)}
e^{-\frac{1}{2}\sigma^2(n-m)^2}
\rho_{nm}.
$$

For a ground-like readout branch, the model uses

$$
\theta = \theta_g,
\qquad
\sigma = \sigma_g.
$$

For an excited-like branch, it uses

$$
\theta = \theta_e,
\qquad
\sigma = \sigma_e.
$$

---

### 3. Calibration convention for the fitted angles

The fitted angles `theta_g_deg` and `theta_e_deg` were measured experimentally from the sequence

$$
\text{readout}
\rightarrow
400\ \mathrm{ns}\ \text{delay}
\rightarrow
\text{conditional } \pi
\rightarrow
\text{Wigner}.
$$

Therefore they are interpreted as total effective cavity backaction angles for the full readout-and-reset sequence.

Because of this, the explicit `reset_delay = 0.42` channel should not add another dispersive cavity rotation.  
Otherwise the cavity rotation during the 400 ns delay would be counted twice.

In the code, the reset-delay channel inside `make_effective_readout` is therefore

```python
reset_delay_channel = make_idle(...)
```

not

```python
reset_delay_channel = make_idle_with_drift(...)
```

The role of this explicit reset-delay channel is to model qubit decay before the conditional pi pulse, not to add an additional calibrated cavity rotation.

---

### 4. Hidden excited-state decay during the readout

If the qubit starts in $|e\rangle$, it can decay during the readout pulse.

The no-jump probability is

$$
p_\mathrm{no\,jump}
=
e^{-T_\mathrm{ro}/T_1^q}.
$$

The jump probability is divided into a few hidden time bins.  
For a branch where the qubit spends a fraction $f_i$ of the readout in $|e\rangle$, the effective readout rotation is approximated as

$$
\theta_i
=
f_i\theta_e + (1-f_i)\theta_g,
$$

and similarly,

$$
\sigma_i
=
f_i\sigma_e + (1-f_i)\sigma_g.
$$

The hidden branches are not sampled explicitly.  
They are summed into four unnormalized blocks:

$$
\tilde\rho_{\mathrm{rec}\,g,\mathrm{true}\,g},
\qquad
\tilde\rho_{\mathrm{rec}\,g,\mathrm{true}\,e},
$$

$$
\tilde\rho_{\mathrm{rec}\,e,\mathrm{true}\,g},
\qquad
\tilde\rho_{\mathrm{rec}\,e,\mathrm{true}\,e}.
$$

This keeps track of both:

1. the recorded classical bit,
2. the physical qubit state after the readout.

---

### 5. Assignment errors

Classical readout assignment errors are included as

$$
p_{g\to e}
=
P(\mathrm{record\ excited}\mid \mathrm{qubit\ ground}),
$$

$$
p_{e\to g}
=
P(\mathrm{record\ ground}\mid \mathrm{qubit\ excited}).
$$

The probabilities of the recorded outcomes are

$$
P(\mathrm{rec}\,g)
=
\mathrm{Tr}
\left[
\tilde\rho_{\mathrm{rec}\,g,\mathrm{true}\,g}
+
\tilde\rho_{\mathrm{rec}\,g,\mathrm{true}\,e}
\right],
$$

$$
P(\mathrm{rec}\,e)
=
\mathrm{Tr}
\left[
\tilde\rho_{\mathrm{rec}\,e,\mathrm{true}\,g}
+
\tilde\rho_{\mathrm{rec}\,e,\mathrm{true}\,e}
\right].
$$

Only the recorded bit is sampled.

---

### 6. Reset delay and conditional pi pulse

After the recorded bit is sampled, the model reconstructs the conditional full qubit+cavity state.  
Then it applies a physical reset delay of duration

$$
t_\mathrm{reset} = 0.42\ \mu\mathrm{s}.
$$

Because the measured angles already include the deterministic cavity rotation of this delay, this reset-delay channel is dissipative-only:

$$
\mathcal L_\mathrm{reset}
=
\frac{1}{T_1^q}\mathcal D[\sigma_-]
+
\frac{1}{2T_\phi^q}\mathcal D[\sigma_z]
+
\frac{1}{T_1^a}\mathcal D[a].
$$

Then the conditional reset is applied:

$$
\rho
\mapsto
X_\pi \rho X_\pi^\dagger
\quad
\text{only if the recorded bit was excited}.
$$

This naturally includes the bad reset path:

$$
\mathrm{recorded}\ e
\rightarrow
e \to g \text{ during } t_\mathrm{reset}
\rightarrow
X_\pi
\rightarrow
g \to e.
$$

This is the main mechanism that should make the neural network prefer readout outcomes where the qubit is recorded in ground, without adding an artificial cost.

---

### 7. Idle and RNN control

After the delayed reset, the state idles for the feedback/storage time:

$$
\rho
\mapsto
e^{t_\mathrm{idle}\mathcal L_\mathrm{idle}}[\rho].
$$

This idle does include the dispersive Hamiltonian

$$
H_\mathrm{idle}
=
\chi a^\dagger a |e\rangle\langle e|
+
\frac{K_a}{2}
a^{\dagger 2}a^2.
$$

Therefore, if the reset accidentally leaves the qubit in $|e\rangle$, the cavity rotates under $\chi$.

The RNN receives only the recorded bit:

$$
h_{k+1},u_k
=
\pi_\Theta(h_k,m_k).
$$

The output $u_k$ defines the feedback-GRAPE controls.

During the feedback-GRAPE pulse, the unitary control chunks already include $H_\mathrm{disp}$ and $H_\mathrm{kerr}$.  
Therefore the interleaved half-idle channels during the pulse are dissipative-only, to avoid double-counting the Hamiltonian.

The reward still uses the cavity fidelity after the full feedback step:

$$
F_{k+1}
=
\langle\psi_\mathrm{target}|
\mathrm{Tr}_q[\rho_{k+1}]
|\psi_\mathrm{target}\rangle.
$$

## Trajectory

````python
def make_trajectory(N_cav, chi, t1_qub, t2_qub, t1_cav, t_idle, psi_target_cav):
    time_step = make_time_step(
        N_cav,
        chi,
        t1_qub,
        t2_qub,
        t1_cav,
        t_idle,
        psi_target_cav,
    )
    
    def trajectory(key, flax_state, params, num_time_steps):
        
        def f(scan_carry, _):
            key, rho, carry = scan_carry
            key, subkey = jax.random.split(key)

            (
                rho,
                carry,
                fid_after_readout,
                fid_after_control,
                meas_out,
                prob,
            ) = time_step(subkey, rho, flax_state, params, carry)

            fid_eff = (
                0.5 * fid_after_readout
                + 0.5 * fid_after_control
            )

            y = (
                fid_eff,
                fid_after_readout,
                fid_after_control,
                meas_out,
                prob,
            )

            scan_carry = key, rho, carry
            return scan_carry, y
        
        key1, key2 = jax.random.split(key)

        carry = init_carry(key1, rec_features, inp_shape)

        psi_init = tensor(basis(2), psi_target_cav)
        rho_init = psi_init @ hconj(psi_init)

        scan_carry = key2, rho_init, carry

        _, ys = jax.lax.scan(
            f,
            scan_carry,
            xs=None,
            length=num_time_steps,
        )

        return ys
    
    return trajectory
````

## Trajectory with reward gradients

````python
def make_trajectory_with_rewgrad(
    N_cav,
    chi,
    t1_qub,
    t2_qub,
    t1_cav,
    t_idle,
    psi_target_cav,
):
    
    trajectory = make_trajectory(
        N_cav,
        chi,
        t1_qub,
        t2_qub,
        t1_cav,
        t_idle,
        psi_target_cav,
    )
    
    def trajectory_with_rewgrad(key, flax_state, num_time_steps):
    
        def func(params):
            traj_out = trajectory(
                key,
                flax_state,
                params,
                num_time_steps,
            )

            (
                fid_eff,
                fid_after_readout,
                fid_after_control,
                meas_out,
                prob,
            ) = traj_out

            eps = 1e-8

            # This is now the objective:
            #
            #     fid_eff = 0.5 * fid_after_readout
            #             + 0.5 * fid_after_control
            #
            # averaged over feedback steps.
            fid_eff_mean = jnp.real(fid_eff.mean())
            fid_eff_mean = jnp.clip(fid_eff_mean, 0.0, 1.0 - eps)

            rew = -jnp.log(1.0 - fid_eff_mean)

            prob_safe = jnp.clip(prob, eps, 1.0)
            log_prob = jnp.log(prob_safe).sum()

            # Pathwise gradient through the differentiable simulation
            # plus REINFORCE term for the sampled measurement outcomes.
            return rew + jax.lax.stop_gradient(rew) * log_prob, traj_out

        grad_func = jax.grad(func, has_aux=True)

        rewgrad, traj_out = grad_func(flax_state.params)

        return rewgrad, traj_out
    
    return trajectory_with_rewgrad
````

## Train on batch

````python
def make_train_on_batch(
        N_cav, chi,
        t1_qub, t2_qub, t1_cav, t_idle,
        psi_target_cav,
        batch_size, num_time_steps
    ):
    
    trajectory_with_rewgrad = make_trajectory_with_rewgrad(
        N_cav, chi, t1_qub, t2_qub, t1_cav, t_idle, psi_target_cav
    )
    
    trajectory_with_rewgrad_vec = jax.vmap(
            trajectory_with_rewgrad, in_axes=(0, None, None)
        )
    
    @jax.jit
    def train_on_batch(key, flax_state):
        keys = jax.random.split(key, batch_size)
        rewgrad, traj_out = trajectory_with_rewgrad_vec(keys, flax_state, num_time_steps)
        lossgrad = jax.tree.map(lambda x: -x.mean(axis=0), rewgrad)
        flax_state = flax_state.apply_gradients(grads=lossgrad) 
        return flax_state, traj_out
    
    return train_on_batch
````

## Test on batch

````python
def make_test_on_batch(
        N_cav, chi,
        t1_qub, t2_qub, t1_cav, t_idle,
        psi_target_cav,
        batch_size, num_time_steps
    ):
    
    trajectory = make_trajectory(
        N_cav, chi, t1_qub, t2_qub, t1_cav, t_idle, psi_target_cav
    )
    
    trajectory_vec = jax.vmap(
            trajectory, in_axes=(0, None, None, None)
        )
    
    @jax.jit
    def test_on_batch(key, flax_state):
        keys = jax.random.split(key, batch_size)
        traj_out = trajectory_vec(keys, flax_state, flax_state.params, num_time_steps)
        return traj_out
    
    return test_on_batch
````

````python
import numpy as np
import jax.numpy as jnp
import jax.scipy.linalg as jsp_linalg
import jax.scipy.special as jsp_special
import matplotlib.pyplot as plt


def coherent_ket_truncated(N, alpha):
    n = jnp.arange(N)
    log_fact = jsp_special.gammaln(n + 1)

    coeffs = jnp.exp(-0.5 * jnp.abs(alpha)**2)
    coeffs = coeffs * (alpha ** n) * jnp.exp(-0.5 * log_fact)

    ket = coeffs.reshape(-1, 1)
    ket = ket / jnp.sqrt((hconj(ket) @ ket).squeeze())

    return ket


def cav_expect_a(rho_cav):
    a = destroy(rho_cav.shape[0])
    return jnp.trace(a @ rho_cav)


def angle_deg(z):
    return np.rad2deg(np.angle(complex(z)))


def angle_diff_deg(z1, z0):
    """
    Wrapped angle difference arg(z1) - arg(z0), in degrees.
    """
    d = np.angle(complex(z1)) - np.angle(complex(z0))
    d = np.angle(np.exp(1j * d))
    return np.rad2deg(d)


def test_disp_vs_readout_rotation_sign(N_cav, chi, alpha0=1.2, t_test=0.42):
    """
    Checks the sign convention of:
        H_disp = chi * n * |e><e|

    against the phenomenological readout rotation:
        rho_nm -> exp(+i theta (n-m)) rho_nm.
    """

    psi_cav = coherent_ket_truncated(N_cav, alpha0)
    rho_cav = psi_cav @ hconj(psi_cav)

    a0 = cav_expect_a(rho_cav)

    # ------------------------------------------------------------
    # 1. Pure dispersive excited-state evolution
    # ------------------------------------------------------------
    a = destroy(N_cav)
    adag = hconj(a)
    n = adag @ a

    H_e = chi * n
    U_e = jsp_linalg.expm(-1j * H_e * t_test)

    rho_disp = U_e @ rho_cav @ hconj(U_e)
    a_disp = cav_expect_a(rho_disp)

    theta_disp_measured = angle_diff_deg(a_disp, a0)
    theta_disp_expected = np.rad2deg(float(-chi * t_test))

    # ------------------------------------------------------------
    # 2. Phenomenological readout rotation with theta = -chi*t
    # ------------------------------------------------------------
    theta = -chi * t_test

    rho_pheno = apply_phase_rotation_and_diffusion(
        rho_cav,
        theta_rad=theta,
        sigma_theta_rad=0.0,
    )

    a_pheno = cav_expect_a(rho_pheno)
    theta_pheno_measured = angle_diff_deg(a_pheno, a0)

    print("chi =", float(chi), "rad/us")
    print("t_test =", t_test, "us")
    print()
    print("Expected H_disp angle -chi*t =", theta_disp_expected, "deg")
    print("Measured H_disp angle        =", theta_disp_measured, "deg")
    print()
    print("Phenomenological theta used  =", np.rad2deg(float(theta)), "deg")
    print("Measured pheno angle         =", theta_pheno_measured, "deg")

    print()
    if theta_disp_measured > 0:
        print("In your simulation, excited-state H_disp rotates in the POSITIVE theta direction.")
    else:
        print("In your simulation, excited-state H_disp rotates in the NEGATIVE theta direction.")

    print()
    print("Your fitted angles:")
    print("theta_g_deg = -18.67")
    print("theta_e_deg = +13.78")

    print()
    print("If the readout-induced rotation is experimentally opposite to excited-state H_disp,")
    print("then the readout-induced angle should have the opposite sign to the measured H_disp angle.")

    # ------------------------------------------------------------
    # Plot centroids
    # ------------------------------------------------------------
    plt.figure(figsize=(4.8, 4.8))

    plt.scatter(
        [np.real(complex(a0))],
        [np.imag(complex(a0))],
        s=90,
        label="initial",
    )

    plt.scatter(
        [np.real(complex(a_disp))],
        [np.imag(complex(a_disp))],
        s=90,
        label="H_disp, qubit excited",
    )

    plt.scatter(
        [np.real(complex(a_pheno))],
        [np.imag(complex(a_pheno))],
        s=60,
        marker="x",
        label=r"pheno $\theta=-\chi t$",
    )

    plt.axhline(0, color="black", linewidth=0.8)
    plt.axvline(0, color="black", linewidth=0.8)

    lim = 1.4 * abs(alpha0)
    plt.xlim(-lim, lim)
    plt.ylim(-lim, lim)
    plt.gca().set_aspect("equal")

    plt.xlabel(r"Re $\langle a\rangle$")
    plt.ylabel(r"Im $\langle a\rangle$")
    plt.grid(alpha=0.3)
    plt.legend()
    plt.title("Dispersive rotation sign check")
    plt.show()


test_disp_vs_readout_rotation_sign(
    N_cav=N_cav,
    chi=chi,
    alpha0=1.2,
    t_test=0.42,
)
````

````python
from tqdm.auto import tqdm
from datetime import datetime
import os
from pathlib import Path

# Normal GRAPE: prepare |g,0> -> |g> tensor psi_target_cav
# Uses the already-defined feedback-GRAPE Hamiltonian through build_drive_htree.

psi_init_grape = tensor(basis(2), basis(N_cav, 0))       # |g,0>
psi_target_grape = tensor(basis(2), psi_target_cav)      # |g> ⊗ target cavity state

grape_optimizer = optax.adam(learning_rate=2e-3)


def grape_fidelity(psi, psi_target):
    overlap = hconj(psi_target) @ psi
    return jnp.abs(overlap[0, 0]) ** 2


def apply_grape_ctrl(grape_ctrl_coeffs):
    e_qub, e_cav = ctrl_from_coeffs(grape_ctrl_coeffs, bsplns_mids)

    # Uses the exact same drift + control Hamiltonians as feedback-GRAPE.
    htree = build_drive_htree(N_cav, chi, e_qub, e_cav)

    psi_t = sesolve_htree(htree, psi_init_grape, t_drive_intvs)
    return psi_t


@jax.jit
def grape_train_step(grape_ctrl_coeffs, grape_opt_state):
    def loss_fn(c):
        psi_t = apply_grape_ctrl(c)
        fid = grape_fidelity(psi_t[-1], psi_target_grape)
        return 1.0 - fid, fid

    (loss_val, fid_val), grads = jax.value_and_grad(loss_fn, has_aux=True)(grape_ctrl_coeffs)
    updates, grape_opt_state = grape_optimizer.update(grads, grape_opt_state)
    grape_ctrl_coeffs = optax.apply_updates(grape_ctrl_coeffs, updates)

    return grape_ctrl_coeffs, grape_opt_state, loss_val, fid_val
````

````python
# ============================================================
# Non-unitary GRAPE helpers
# Same method as the older notebook:
#   decay_fidelity(c) = fidelity after mesolve_htree(htree, c_ops, rho0, time_intervals)
# but using the new notebook Hamiltonian build_drive_htree(...)
# ============================================================

rho_init_grape = psi_init_grape @ hconj(psi_init_grape)

grape_decay_c_ops = build_c_ops(
    N_cav,
    t1_qub,
    t2_qub,
    t1_cav,
)


def decay_grape_fidelity(grape_ctrl_coeffs):
    e_qub, e_cav = ctrl_from_coeffs(grape_ctrl_coeffs, bsplns_mids)

    # Important: this is the Hamiltonian from the new notebook
    htree = build_drive_htree(N_cav, chi, e_qub, e_cav)

    rho_out = mesolve_htree(
        htree,
        grape_decay_c_ops,
        rho_init_grape,
        t_drive_intvs,
    )

    # depending on toolbox version, mesolve_htree may return final rho or trajectory
    rho_fin = rho_out[-1] if rho_out.ndim == 3 else rho_out

    fid = hconj(psi_target_grape) @ rho_fin @ psi_target_grape
    return jnp.real(fid[0, 0])
````

````python
# ============================================================
# Non-unitary GRAPE fine-tuning
# Rerunning this cell continues from the current decay-optimized pulse.
# ============================================================

n_steps_decay = 1000
learning_rate = 2e-7
decay_param_clip = 2.0

decay_optimizer = optax.adam(learning_rate=learning_rate)

try:
    grape_ctrl_coeffs_decay
    decay_opt_state
    grape_decay_fid_hist
    print("Continuing existing non-unitary GRAPE optimization.")
except NameError:
    grape_ctrl_coeffs_decay = jnp.clip(grape_ctrl_coeffs, -decay_param_clip, decay_param_clip)
    decay_opt_state = decay_optimizer.init(grape_ctrl_coeffs_decay)
    grape_decay_fid_hist = []
    print("Starting non-unitary GRAPE from current unitary-GRAPE pulse.")


@jax.jit
def decay_grape_train_step(ctrl_coeffs, opt_state):
    def loss_fn(c):
        fid = decay_grape_fidelity(c)
        return 1.0 - fid, fid

    (loss_value, fid_value), grads = jax.value_and_grad(loss_fn, has_aux=True)(ctrl_coeffs)

    updates, opt_state = decay_optimizer.update(grads, opt_state)
    ctrl_coeffs = optax.apply_updates(ctrl_coeffs, updates)

    # Same idea as the older notebook: keep coefficients bounded.
    ctrl_coeffs = jnp.clip(ctrl_coeffs, -decay_param_clip, decay_param_clip)

    return ctrl_coeffs, opt_state, loss_value, fid_value


print("Initial/current decay fidelity:", float(decay_grape_fidelity(grape_ctrl_coeffs_decay)))

pbar = tqdm(range(n_steps_decay), desc="non-unitary GRAPE")
for _ in pbar:
    grape_ctrl_coeffs_decay, decay_opt_state, decay_loss_val, decay_fid_val = decay_grape_train_step(
        grape_ctrl_coeffs_decay,
        decay_opt_state,
    )

    grape_decay_fid_hist.append(float(decay_fid_val))
    pbar.set_postfix(fid=f"{float(decay_fid_val):.8f}")

# Make downstream cells use the non-unitary-optimized pulse.
grape_ctrl_coeffs = grape_ctrl_coeffs_decay
grape_ctrl_coeffs_np = np.asarray(grape_ctrl_coeffs, dtype=float)

# Keep your existing export cell compatible if it uses grape_fid_hist[-1].
grape_fid_hist.append(float(grape_decay_fid_hist[-1]))

print("Final non-unitary-GRAPE fidelity:", grape_decay_fid_hist[-1])
print("grape_ctrl_coeffs shape:", grape_ctrl_coeffs_np.shape)

plt.figure(figsize=(6, 3))
plt.semilogy(1.0 - np.asarray(grape_decay_fid_hist))
plt.xlabel("non-unitary GRAPE step")
plt.ylabel("infidelity with decoherence")
plt.title("Non-unitary GRAPE fine-tuning")
plt.tight_layout()
plt.show()
````

````python
import jax
import jax.numpy as jnp
import jax.scipy.linalg as jsp_linalg

def measure_sz(
    key,
    rho_init,
    chi=chi,
    t_feedback=0.42,
    readout_duration=1.6,
    theta_g_deg=-18.67,
    theta_e_deg=+13.78,
):
    """
    Projective qubit measurement + experimentally calibrated readout-induced
    cavity rotation.

    Convention:
        outcome_indx = 0  -> qubit ground    -> outcome = +1
        outcome_indx = 1  -> qubit excited   -> outcome = -1

    Readout-induced cavity rotations:
        qubit ground  : theta_g_deg = -18.67 deg, clockwise
        qubit excited : theta_e_deg = +13.78 deg, anticlockwise

    Positive theta means anticlockwise rotation in phase space.
    """

    # Measurement probabilities: [p_ground, p_excited]
    probs = rho_init.diagonal().reshape(2, -1).sum(axis=1).real

    outcome_indx = jax.random.choice(key, 2, p=probs)
    prob = probs[outcome_indx]

    N_cav = rho_init.shape[-1] // 2

    # Select the cavity block corresponding to the measurement outcome
    rho_cav = jnp.where(
        outcome_indx == 0,
        rho_init[:N_cav, :N_cav],
        rho_init[N_cav:, N_cav:],
    )

    # ------------------------------------------------------------
    # Readout-induced cavity rotation, conditioned on qubit outcome
    # ------------------------------------------------------------
    theta_deg = jnp.where(
        outcome_indx == 0,
        theta_g_deg,
        theta_e_deg,
    )

    theta = theta_deg * jnp.pi / 180.0

    a_c = destroy(N_cav)
    adag_c = hconj(a_c)
    n_c = adag_c @ a_c

    # Positive theta gives anticlockwise Wigner rotation.
    U_ro = jsp_linalg.expm(1j * theta * n_c)

    rho_cav = U_ro @ rho_cav @ hconj(U_ro)

    # Qubit reset to ground
    rho_fin_qub = basis(2) @ hconj(basis(2))
    rho_fin = tensor(rho_fin_qub, rho_cav)

    # Renormalize after projection
    rho_fin = rho_fin / jnp.trace(rho_fin)

    # Outcome definition:
    # ground -> +1, excited -> -1
    outcome = 1 - 2 * outcome_indx

    return outcome, prob, rho_fin
````
